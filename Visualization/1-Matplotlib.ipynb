# **Matplotlib:-**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,200)
x
```

```python
y=np.sin(x)
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("Title")
plt.show()
```

## **"--":-**
is used to break the graph and b is used to graph colour**

```python
plt.figure(figsize=(12,4))
plt.plot(x,y,"--r")
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("Title")
plt.grid()
plt.show()
```

# **Scatter:-**

## **alpha:-**
### It is reprsenting the intenscity 

```python
x=np.random.rand(50)
y=np.random.rand(50)
colour=np.random.rand(50)
size=1000*np.random.rand(50)
plt.scatter(x,y,c=colour,s=size,alpha=.1)
plt.show()
```

# **Subplots:-**

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)# This is used to determine the block of graph at matrix level
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)# This function is take 3 argument, first is row and second is column and third is give figure number
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)

plt.show()
```
