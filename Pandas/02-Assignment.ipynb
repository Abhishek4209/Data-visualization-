# **Q11:-** 
## **You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g.Monday, Tuesday) corresponding to each date in the 'Date' column.**

#### **For example, if df contains the following values:**

## **Date**
### 0 2023-01-01
### 1 2023-01-02
### 2 2023-01-03
### 3 2023-01-04
### 4 2023-01-05
## **Your function should create the following DataFrame:**

## **Date Weekday**
### 0 2023-01-01 Sunday
### 1 2023-01-02 Monday
### 2 2023-01-03 Tuesday
### 3 2023-01-04 Wednesday
### 4 2023-01-05 Thursday
## **The function should return the modified DataFrame.**

```python
import pandas as pd

df = pd.DataFrame(data={"Date":["2023-01-01"
,"2023-01-02","2023-01-03","2023-01-04","2023-01-05","2023-01-06","2023-01-07","2023-01-08","2023-01-09"]})
df['Date'] = pd.to_datetime(df['Date'])
df['Weekday'] = df['Date'].dt.dayofweek

days = {0:'Sunday',1:'Monday',2:'Tuesday',3:'Wednesday',4:'Thursday',5:'Friday',6:'Saturday'}

df['Weekday'] = df['Weekday'].apply(lambda x: days[x])
df
```

## **write a Python function to select all rows where the date is between '2023-01-01' and '2023-01-31' given a Pandas DataFrame df with a column 'Date' that contains timestamps, '.**

```python
import datetime
given_date=[]
start = datetime.datetime.strptime("1-01-2023", "%d-%m-%Y")
end = datetime.datetime.strptime("15-07-2023", "%d-%m-%Y")
date_generated = [start + datetime.timedelta(days=x) for x in range(0, (end-start).days)]

for date in date_generated:
    a=(date.strftime("%d-%m-%Y"))
    given_date.append(a)
```
