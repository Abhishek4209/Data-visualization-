# **Merge;-**

```python
data1={"key1":[1,2,3,4,5],"key2":[2,3,1,4,5],"key3":[3,2,1,4,5]}
data2={"key1":[1,2,31,4,54],"key4":[22,33,11,43,53],"key5":[32,21,12,43,54]}
```

```python
a=pd.DataFrame(data1)
a
```

```python
b=pd.DataFrame(data2)
b
```

```python
pd.merge(a,b)
```

```python
pd.merge(a,b,how="left")
```

```python
pd.merge(a,b,how="right")
```

```python
pd.merge(a,b,how="outer",on="key1")
```

```python
pd.merge(a,b,how="cross")
```
